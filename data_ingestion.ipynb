```python
from dotenv import load_dotenv
import os

load_dotenv(
    dotenv_path=".env",  # custom path
    override=True        # overwrite existing env vars
)
```

### Creating catalog, schema and volume for Ingestion Data

```python

catalog_name = os.getenv("CATALOG")
schema_name = os.getenv("SCHEMA")
volume_name = os.getenv("VOLUME")

print(f"{catalog_name}, {schema_name}, {volume_name}") 
```

```python
spark.sql(f"CREATE CATALOG IF NOT EXISTS {catalog_name}")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {catalog_name}.{schema_name}")
spark.sql(f"CREATE VOLUME IF NOT EXISTS {catalog_name}.{schema_name}.{volume_name}")

print("Catalog, schema, and volume created successfully ✅")
```

### creaing folder inside the volume

```python
os.mkdirs(f"/Volumes/{catalog_name}/{schema_name}/{volume_name}/global_fashion", exist_ok=True)
```

### Ingesting the data using Kaggle API

```python
from kaggle.api.kaggle_api_extended import KaggleApi

api = KaggleApi()
api.authenticate()

volume_path = "/Volumes/{catalog_name}/{schema_name}/{volume_name}/global_fashion"
kaggle_path = "ricgomes/global-fashion-retail-stores-dataset"

api.dataset_download_files(
    kaggle_path,
    path=volume_path,
    unzip=True
)
```
